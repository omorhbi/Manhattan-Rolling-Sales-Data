```python
cleaned_data["GROSS SQUARE FEET"] = cleaned_data["GROSS SQUARE FEET"].str.replace(',', '')
cleaned_data["LAND SQUARE FEET"] = cleaned_data["LAND SQUARE FEET"].str.replace(',', '')
cleaned_data["SALE PRICE"] = cleaned_data["SALE PRICE"].str.replace(',', '')
```

### 6. Describe the kind of buildings that were sold

```python
# get the building classes
print(cleaned_data["BUILDING CLASS CATEGORY"].unique())
```

```python
# clean the category of numbers since they're not necessary
cleaned_data["BUILDING CLASS CATEGORY"] = cleaned_data["BUILDING CLASS CATEGORY"].str.replace('\d+','')
cleaned_data["BUILDING CLASS CATEGORY"] = cleaned_data["BUILDING CLASS CATEGORY"].str.strip()
```

```python
# Do this to locate one family houses and office buildings
building_class_count = cleaned_data["BUILDING CLASS CATEGORY"].value_counts()
building_class_count = building_class_count.sort_index()
building_class_count # helps locate the indices 
```

```python
# extract the counts for one family houses and office buildings have while subtracting them from the total for others
one_family_houses = building_class_count[27]
office_buildings = building_class_count[26]
condos = building_class_count[0] + building_class_count[2] + building_class_count[5:18].sum() + building_class_count[35]
others = building_class_count.sum() - office_buildings - one_family_houses - condos

# set up a dictionary so that when it comes to making the bar graph, it'll automatically use the key and values of the dict
data_dictionary = {'One Family Houses': one_family_houses, 'Office Buildings': office_buildings, 'Condos': condos, 'Others': others}
```

```python
# make the series
building_class_data = pd.Series(data_dictionary)
data_dictionary
```

```python
# create the graph
building_class_data.plot(kind = 'bar', figsize = (8, 8), rot = 0, title = "Building Class Categories")
```

Since series doesn't seem to have labels for x and y axis, the y-axis is for properties and the x-axis is for the building categories

### 7. Calculate summary statistics for the prices of properties sold for all of Manhattan and for a couple of select neighborhoods

```python
# Calculating overall prices for the entire data set
property_prices = cleaned_data["SALE PRICE"]
price_mean = round(property_prices.mean(), 2)
price_median = int(property_prices.median())
price_max = property_prices.max()
price_min = property_prices.min()
first_quartile = property_prices.quantile(q = 0.25).astype(int)
third_quartile = property_prices.quantile(q = 0.75).astype(int)
property_price_dictionary = {"Mean": price_mean, "Median": price_median, "Max": price_max, "Min": price_min, 
                             "First Quartile": first_quartile, "Third Quartile": third_quartile}
```

```python
#property_price_dictionary
price_data = pd.DataFrame(data = property_price_dictionary, index = ["Manhattan Properties"])
price_data
```

```python
neighborhood_count # this is to help locate the indices for the neighborhoods
```

```python
# filter out the rows that contain Greenwich Central and East Village in their respective variables
greenwich_central = cleaned_data[cleaned_data["NEIGHBORHOOD"] == "GREENWICH VILLAGE-CENTRAL"]
east_village = cleaned_data[cleaned_data["NEIGHBORHOOD"] == "EAST VILLAGE"]
```

```python
# do calculations for Greenwich Village Central
gw_central_prices = greenwich_central["SALE PRICE"]
gw_mean = round(gw_central_prices.mean(), 2)
gw_median = int(gw_central_prices.median())
gw_max = gw_central_prices.max()
gw_min = gw_central_prices.min()
gw_first_quartile = gw_central_prices.quantile(q = 0.25).astype(int)
gw_third_quartile = gw_central_prices.quantile(q = 0.75).astype(int)
gw_dictionary = {"Mean": gw_mean, "Median": gw_median, "Max": gw_max, "Min": gw_min, "First Quartile": gw_first_quartile, 
                 "Third Quartile": gw_third_quartile}
```

```python
# do calculations for East Village
ev_prices = east_village["SALE PRICE"]
ev_mean = round(ev_prices.mean(), 2) # round to 2 decimal points
ev_median = int(ev_prices.median()) # for some reason median returns a float initially
ev_max = ev_prices.max()
ev_min = ev_prices.min()
ev_first_quartile = ev_prices.quantile(q = 0.25).astype(int)
ev_third_quartile = ev_prices.quantile(q = 0.75).astype(int)
ev_dictionary = {"Mean": ev_mean, "Median": ev_median, "Max": ev_max, "Min": ev_min, "First Quartile": ev_first_quartile, 
                 "Third Quartile": ev_third_quartile}
```

```python
# pass in the two dictionaries as a list of dicts for the data argument
ev_gw_dataframe = pd.DataFrame(data = [gw_dictionary, ev_dictionary], index = ["Greenwich Village", "East Village"])
```
